```python
# Import scrapy
import scrapy

# Import the CrawlerProcess: for running the spider
from scrapy.crawler import CrawlerProcess
```

```python
# Create the Spider class

url = "https://archive.ics.uci.edu/ml/index.php"
links = []

class UCI_ML_Spider(scrapy.Spider):
    
    name = "uci_ml_spider"
  
    # start_requests method
    def start_requests(self):
    
        yield scrapy.Request(url = url,
                         callback = self.parse_front)
    
    
    # First parsing method
    def parse_front(self, response):
        
        links_to_follow = response.css('span.normal > a::attr(href)').extract()
        links_to_follow = links_to_follow[7:-5]
        links = ['https://archive.ics.uci.edu/ml/' + str(l) for l in links_to_follow]
        
        for url in links:
            yield response.follow(url = url,
                                  callback = self.parse_pages)
            
    # Second parsing method
    def parse_pages(self, response):
        
        data_title = response.xpath('//span[contains(@class, "heading")]/b/text()').extract()
        
    
        data_desc = response.xpath('//p[contains(@class, "normal")]/text()').extract()
        
        for data_title, data_desc in zip(data_title, data_desc):
            uci_dict[data_title] = data_desc
  
```

```python
# Initialize the dictionary **outside** of the Spider class
uci_dict = dict()
```

```python
# Run the Spider
process = CrawlerProcess()
process.crawl(UCI_ML_Spider)
process.start()
```
